```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("postgresql://postgres:1205@localhost:5432/fintech_affluent")

prospects = pd.read_csv("C:\\Users\\notya\\Desktop\\Yash\\Data Projects\\fintech-affluent-acquisition\\data\\raw\\prospects.csv")

prospects['converted'] = prospects['converted'].astype(bool)

prospects.to_sql(
    "prospects_raw",
    engine,
    schema="raw",
    if_exists="append",
    index=False
)

print("Prospects loaded successfully.")
```
