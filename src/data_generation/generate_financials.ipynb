```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine

engine = create_engine("postgresql://postgres:1205@localhost:5432/fintech_affluent")

# Fetch customers + income
query = """
SELECT c.customer_id, p.annual_income
FROM analytics.customers c
JOIN raw.prospects_raw p
ON c.prospect_id = p.prospect_id
"""

df = pd.read_sql(query, engine)

np.random.seed(42)

multiplier = df["annual_income"] * np.random.uniform(1, 3, len(df))

financials = pd.DataFrame({
    "customer_id": df["customer_id"],
    "savings_balance": multiplier * 0.2,
    "fixed_deposits": multiplier * 0.25,
    "investments": multiplier * 0.3,
    "insurance_aum": multiplier * 0.1,
    "demat_holdings": multiplier * 0.15
})

financials["total_trb"] = financials.iloc[:,1:].sum(axis=1)

financials.to_sql(
    "financials",
    engine,
    schema="analytics",
    if_exists="append",
    index=False
)

print("Financials inserted successfully.")
```
