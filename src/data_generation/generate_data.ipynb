```python
import numpy as np
import pandas as pd
from faker import Faker

np.random.seed(42)
fake = Faker()

N = 50000

# -------------------------
# Generate Prospect Data
# -------------------------

ages = np.random.randint(25, 60, N)

annual_income = np.random.normal(1200000, 500000, N)
annual_income = np.clip(annual_income, 300000, None)

occupation = np.random.choice(
    ["Salaried", "Business Owner", "Professional", "Self-Employed"],
    N,
    p=[0.5, 0.2, 0.2, 0.1]
)

city_tier = np.random.choice([1,2,3], N, p=[0.5,0.3,0.2])

credit_score = np.random.normal(720, 60, N)
credit_score = np.clip(credit_score, 500, 850)

digital_score = np.random.randint(1, 100, N)

campaign_channel = np.random.choice(
    ["Digital Ads", "Relationship Manager", "Branch Walk-in", "Referral"],
    N,
    p=[0.4, 0.2, 0.25, 0.15]
)

acquisition_cost = np.random.normal(3000, 500, N)
acquisition_cost = np.clip(acquisition_cost, 1500, None)

# -------------------------
# Conversion Probability
# -------------------------

conversion_prob = (
    (annual_income / 2000000) * 0.3 +
    (credit_score / 850) * 0.3 +
    (digital_score / 100) * 0.1 +
    (1 / city_tier) * 0.2
)

conversion_prob = np.clip(conversion_prob, 0, 0.8)
converted = np.random.binomial(1, conversion_prob)

prospects = pd.DataFrame({
    "age": ages,
    "annual_income": annual_income,
    "occupation": occupation,
    "city_tier": city_tier,
    "credit_score": credit_score,
    "digital_score": digital_score,
    "campaign_channel": campaign_channel,
    "acquisition_cost": acquisition_cost,
    "converted": converted
})

prospects.to_csv("C:\\Users\\notya\\Desktop\\Yash\\Data Projects\\fintech-affluent-acquisition\\data\\raw\\prospects.csv", index=False)

print("Prospects generated:", len(prospects))
```
